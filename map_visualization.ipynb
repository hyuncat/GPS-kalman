```python
import folium
from folium.plugins import HeatMap
import pandas as pd
```

```python
# Read person10's filtered data from csv
person10_kalman = pd.read_csv("data/kalman_filtered/kalman_person10_2008-06-13.csv")
person10_kalman
```

```python
def plot_walking_heatmap(kalman_df):
    folium_map = folium.Map(location=[kalman_df['lat_filtered'].mean(), kalman_df['long_filtered'].mean()], 
                            zoom_start=14)
    HeatMap(data=kalman_df[['lat', 'long']], gradient={0.4: 'blue', 0.65: 'cyan'}, blur=6, radius=10).add_to(folium_map)
    HeatMap(data=kalman_df[['lat_filtered', 'long_filtered']], gradient={0.4: 'red', 0.65: 'yellow'}, blur=5, radius=5).add_to(folium_map)
    return folium_map
```

```python
plot_walking_heatmap(person10_kalman)
```

```python
person10_b_kalman = pd.read_csv("data/kalman_filtered/kalman_person10_2008-06-18.csv")
plot_walking_heatmap(person10_b_kalman)
```
