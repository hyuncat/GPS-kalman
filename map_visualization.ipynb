```python
import folium
from folium.plugins import HeatMap
import pandas as pd
```

```python
# Read person10's filtered data from csv
person10_kalman = pd.read_csv("data/kalman_filtered/kalman_person10_2008-06-13.csv")
person10_kalman
```

```python
folium_map2 = folium.Map(location=[person10_kalman['lat_filtered'].mean(), person10_kalman['long_filtered'].mean()], 
                        zoom_start=14)
HeatMap(data=person10_kalman[['lat', 'long']], gradient={0.4: 'blue', 0.65: 'cyan'}, blur=6, radius=10).add_to(folium_map2)
HeatMap(data=person10_kalman[['lat_filtered', 'long_filtered']], gradient={0.4: 'red', 0.65: 'yellow'}, blur=5, radius=5).add_to(folium_map2)
folium_map2
```
